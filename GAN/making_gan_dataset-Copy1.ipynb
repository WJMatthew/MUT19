```python
def crop_image(im, location, size, pid='test'):
    #location, size = {}, {}
    #location['x'] = 320
    #location['y'] = 647

    left = location['x']
    top = location['y']
    right = location['x'] + size['width']
    bottom = location['y'] + size['height']

    im = im.crop((left, top, right, bottom)) # defines crop points
    fn = f'cropped_{pid}_test.png'
    im.save(fn)
    return im, fn
```
