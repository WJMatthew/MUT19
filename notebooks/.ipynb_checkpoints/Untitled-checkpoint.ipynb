```python
temp_list = []
temp_list2 = []

for _, name in zip(tmp.values, new_df['Name'].values):
    p = []
    temp_ = []
    
    for i, s in enumerate(_):
        #x = re.findall(r'\d{4}-\d{4}', s)
        y = re.findall(r'\d{4}', s)
        if len(y) == 1: # single year
            y.append(y[0])
        z = re.findall(r'\b[^\W\d_]+\b', s)
        #print(x)
        #print(y)
        #print(z)
        
        temp_.append(pd.DataFrame({'Team': z[0], 'Year0': y[0], 'yearN': y[1] }, index=[0]))
        
        p.append({z[0] : y})
    
    temp_list2.append(p)
    temp_list.append(temp_)
```

```python
def team_checker(team_dict):
    teams_list = []
    
    for _ in team_dict:
        if ( (_=='LA') & ( (int(team_dict.get(_)[0]) <= 1994) | (int(team_dict.get(_)[0]) >= 2015))): # Rams
            #print('# RAMS')
            teams_list.append('RAMS')
            
        elif (_=='Bal'):
            if int(team_dict.get(_)[0]) <= 1995: # Ravens
                #print('# Colts')
                teams_list.append('Ind')
            else:
                #print('# Ravens')
                teams_list.append('Bal')
        
        elif (_=='Hous'):
            if int(team_dict.get(_)[0]) <= 1996: # Ravens
                #print('# Oilers (Titans)')
                teams_list.append('Ten')
            else:
                #print('# Texans')
                teams_list.append('Hou')

        elif (_=='Pho'):
            #print('# Ari')
            teams_list.append('Ari')
            
        else: 
            teams_list.append(_)
        
    return teams_list[0]
```

```python
#print( type(temp_list2[0][0]))
names = new_df['Name'].values

p_dict = {}
for name, thing in zip(names, temp_list2):
    team_list = []
    for _ in thing:
        #print(_)
        team_list.append(team_checker(_))
    p_dict.update({name.strip(): '|'.join(team_list)})
```
